```python
import numpy as np

def sigmoid(x: np.array) -> np.array:
    return 1 / (1 + np.exp(-x))

def d_sigmoid(x: np.array) -> np.array:
    return sigmoid(x) * (1 - sigmoid(x))

def d_tanh(x: np.array) -> np.array:
    return 1 - np.tanh(x) ** 2

def relu(x: np.array) -> np.array:
    return np.maximum(0, x)

def d_relu(x: np.array) -> np.array:
    return np.where(x > 0, 1, 0)

def linear(x: np.array) -> np.array:
    return x

def d_linear(x: np.array) -> np.array:
    return np.ones(x.shape)

def return_activations(name_haf:str, name_oaf:str):
    activations = {}
    # Hidden activate function
    if name_haf == 'sigmoid':
        activations['hidden_activation'] = sigmoid
        activations['derivative_hidden_activation'] = d_sigmoid
    elif name_haf == 'relu':
        activations['hidden_activation'] = relu
        activations['derivative_hidden_activation'] = d_relu
    elif name_haf == 'tanh':
        activations['hidden_activation'] = np.tanh
        activations['derivative_hidden_activation'] = d_tanh
    elif name_haf == 'linear':
        activations['hidden_activation'] = linear
        activations['derivative_hidden_activation'] = d_linear

    # Output activate function
    if name_oaf == 'sigmoid':
        activations['output_activation'] = sigmoid
        activations['derivative_output_activation'] = d_sigmoid
    elif name_oaf == 'relu':
        activations['output_activation'] = relu
        activations['derivative_output_activation'] = d_relu
    elif name_oaf == 'tanh':
        activations['output_activation'] = np.tanh
        activations['derivative_output_activation'] = d_tanh
    elif name_oaf == 'linear':
        activations['output_activation'] = linear
        activations['derivative_output_activation'] = d_linear
    
    return activations
```

```python
if __name__ == '__main__':
    from matplotlib import pyplot as plt
    x = np.arange(-10, 10, 0.1)
    # sigmoid
    y = sigmoid(x)
    dy = d_sigmoid(x)
    plt.figure(figsize=(10, 4))
    plt.subplot(121)
    plt.plot(x, y)
    plt.subplot(122)
    plt.plot(x, dy)
    # tanh
    y = np.tanh(x)
    dy = d_tanh(x)
    plt.figure(figsize=(10, 4))
    plt.subplot(121)
    plt.plot(x, y)
    plt.subplot(122)
    plt.plot(x, dy)
    # relu
    y = relu(x)
    dy = d_relu(x)
    plt.figure(figsize=(10, 4))
    plt.subplot(121)
    plt.plot(x, y)
    plt.subplot(122)
    plt.plot(x, dy)
    # relu
    y = linear(x)
    dy = d_linear(x)
    plt.figure(figsize=(10, 4))
    plt.subplot(121)
    plt.plot(x, y)
    plt.subplot(122)
    plt.plot(x, dy)
    plt.show()
```
